```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
counts = [250, 600, 300, 350, 200]

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(categories, counts, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('Distribution of Categories')
plt.show()
```

```python
import numpy as np
import seaborn as sns

# Generate some random ages (example data)
np.random.seed(0)
ages = np.random.randint(20, 60, size=100)  # Generate 100 random ages between 20 and 60

# Plotting
plt.figure(figsize=(8, 5))
sns.histplot(ages, bins=10, kde=False, color='salmon')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Ages')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for genders
genders = ["Male", "Female", "Female", "Male", "Male", "Female", "Female", "Male", "Female", "Male", "Male", "Female", "Female", "Male", "Female", "Female", "Male", "Female", "Female", "Male", "Female"]

# Counting occurrences
gender_counts = {
    "Male": genders.count("Male"),
    "Female": genders.count("Female")
}

# Creating bar chart
plt.bar(gender_counts.keys(), gender_counts.values(), color=['blue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
Gender Distribution: The bar chart shows the count of passengers by gender. From the Titanic dataset, we observe that there were more males than females aboard.

Age Distribution: The histogram displays the distribution of ages among passengers. It indicates that the majority of passengers were in their 20s and 30s, with a few outliers above 60 years.
```
